```python
#Part 1 - Basic Image Loading and Manipulation
import cv2
import numpy as np
import matplotlib.pyplot as plt
dog = cv2.imread('./images/dog.jpg')
'''Convert OpenCV (B,G,R) imread to matplotlib (R,G,B)'''
b,g,r = cv2.split(dog)
dog = cv2.merge([r,g,b])
plt.imshow(dog)
plt.title("Visualize Image")
plt.show()

#Info about image
print(dog.shape)
print("Height = " + str(dog.shape[0]) + ' pixels')
print("Width = " + str(dog.shape[1]) + ' pixels')
print("Resolution = " + str(dog.shape[0] * dog.shape[1]) + ' pixels')
print("Total Pixels in image = " + str(dog.shape[0]*dog.shape[1]*dog.shape[2]))

#Individual Pixels
height, width = dog.shape[0], dog.shape[1]
R,G, B = dog[0,0]
print("First Pixel")
print(R,G,B)
R, G, B = dog[int(height/2),int(width/2)]
print("Pixel in the Center of the Image")
print(R,G,B)
R,G,B = dog[height-1, width-1]
print("Last Pixel:")
print(R,G,B)
```

```python
#Visualize R,G,B components of image
filler = np.zeros((383,313), np.uint8)
Red, Green, Blue = cv2.split(dog)
plt.imshow(Red, cmap = "gray")
plt.title("Red Component")
plt.show()
plt.imshow(Green, cmap = "gray")
plt.title("Green Component")
plt.show()
plt.imshow(Blue, cmap = "gray")
plt.title("Blue Component")
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
#Draw Rectangle on the Image
example = np.zeros((512,512,3), np.uint8)
cv2.rectangle(example, (100, 100), (300, 250), (255, 255, 255), -1)
#plt.imshow(example, cmap='gray', interpolation = 'bicubic')
#plt.show()
RectangleDog = cv2.imread('./images/dog.jpg')
b,g,r = cv2.split(RectangleDog)
RectangleDog = cv2.merge([r,g,b])

#LOOP THROUGH IMAGE AND DRAW RECTANGLE AROUND THE DOG'S FACE
for i in range(15, 125):
    for j in range(70, 200):
        if ((i >= 15 and i <= 20) or (i >= 120 and i <= 125)):
                RectangleDog[i][j][0] = 0
                RectangleDog[i][j][1] = 0
                RectangleDog[i][j][2] = 0
for i in range(15, 125):
    for j in range(70, 200):
        if ((j >= 70 and j <= 75) or (j >= 195 and j <= 200)):
                RectangleDog[i][j][0] = 0
                RectangleDog[i][j][1] = 0
                RectangleDog[i][j][2] = 0
plt.imshow(RectangleDog)
plt.show()
```
